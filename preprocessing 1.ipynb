**Function to pull the data of single student**






```
def pull_id(student_id):
    save_path = "/content/drive/My Drive/CNN/training_data/"
    s = pd.DataFrame()
    s = df014.loc[df014['id_student'] == student_id]
    # s = s.sort_values('date')
    # s = s.groupby(['id_student','num_of_prev_attempts','code_module','code_presentation'])
    s = s[['id_student','code_module','date','sum_click_y','num_of_prev_attempts','code_presentation','final_result']]
    s = s.sort_values('date')
    code_list  = s.code_module.unique()
    for i in code_list:
      class_data = class_mod(i,s) 
      presentation = class_data.code_presentation.unique()
      for j in presentation:
        presentation_data = pres_data(j,class_data)
        previous = presentation_data.num_of_prev_attempts.unique()
        for k in previous:
          plot_data = previous_data(k,presentation_data)
          temp = plot_data['final_result'].unique()
          stdresult = temp[0]
          plot_ts(save_path,student_id,i,j,k,'final_result',stdresult,plot_data)
    
```

**Function to plot a graph**

```
import matplotlib.pyplot as plt

def plot_ts(save_path,id_student,code_module,code_presentation,num_of_prev_attempts,target1,value1,a):
  string = '{}_{}_{}_{}_{}_{}.png'.format(id_student,code_module,code_presentation,num_of_prev_attempts,target1,value1)
  if (string in files):
    print('file is present')
  else:
    axis = a.plot(subplots=True,legend=False, figsize=(10,5),x='date', y='sum_click_y', kind='line')
    for ax in axis:
    
      plt.xlabel("")
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    #ax.yticks(color='w')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    path = '{}/{}_{}_{}_{}_{}_{}.png'.format(save_path,id_student,code_module,code_presentation,num_of_prev_attempts,target1,value1)
    plt.savefig(path)
    print(path)
```

```
def class_mod(module,data):
  return_data = data.loc[data['code_module']== module]
  return(return_data)
```

```
def pres_data(press,data):
  return_data = data.loc[data['code_presentation']== press]
  return(return_data)
```

```
def previous_data(prev,data):
  return_data = data.loc[data['num_of_prev_attempts']== prev]
  return(return_data)
```
